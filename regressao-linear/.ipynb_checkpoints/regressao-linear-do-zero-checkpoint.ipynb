```python
# Imports
import sys
from datetime import datetime

from numpy import *

import altair as alt
import pandas as pd

alt.renderers.enable('notebook')
```

```python
def compute_error_for_given_points(b, m, points):
    total_error = 0
    for i in range(len(points)):
        x = points[i, 0]
        y = points[i, 1]
        total_error += (y - (m * x + b)) ** 2
        
    return total_error / float(len(points))
```

```python
def step_gradient(b_current, m_current, points, learning_rate):
    # gradiente descendente
    b_gradient = 0
    m_gradient = 0
    n = float(len(points))
    
    for i in range(len(points)):
        x = points[i, 0]
        y = points[i, 1]
        b_gradient += -(2/n) * (y - ((m_current * x) + b_current))
        m_gradient += -(2/n) * x * (y - ((m_current * x) + b_current))
        
    new_b = b_current - (learning_rate * b_gradient)
    new_m = m_current - (learning_rate * m_gradient)
    
    return [new_b, new_m]
        
```

```python
def gradient_descent_runner(points, starting_b, starting_m, num_iterations, learning_rate, printar_rss, plot_rss):
    b = starting_b
    m = starting_m
    
    iteracao_rss = {"iteracao": [], 
        "rss": []}
    
    for i in range(num_iterations):
        points_array = array(points)
        b, m = step_gradient(b, m, points_array, learning_rate)
        
        if printar_rss:
            print "iteracao: " + str(i + 1) + ", b:  " + str(b) + ", m: " + str(m) + ", erro: " + str(compute_error_for_given_points(b, m, points_array))
        
        iteracao_rss["iteracao"].append(i + 1)
        iteracao_rss["rss"].append(compute_error_for_given_points(b, m, points_array))
    
    if plot_rss:
        return [b, m, iteracao_rss]
    else:
        return [b, m]
```

```python
def run(file, num_iterations = 1000, learning_rate = 0.0001, printar_rss = False, plot_rss = False):
    points = genfromtxt(file, delimiter=",")
    
    # y = mx + b (fórmula de inclinação da reta)
    initial_b = 0
    initial_m = 0
    
    return gradient_descent_runner(points, initial_b, initial_m, num_iterations, learning_rate, printar_rss, plot_rss)
```

```python
[b, m] = run("data/data.csv")

print "b: " + str(b)
print "m: " + str(m)
```

## Verificando o tamanho do gradiente ao longo das iterações

```python
def step_gradient(b_current, m_current, points, learning_rate):
    # gradiente descendente
    b_gradient = 0
    m_gradient = 0
    n = float(len(points))
    
    for i in range(len(points)):
        x = points[i, 0]
        y = points[i, 1]
        b_gradient += -(2/n) * (y - ((m_current * x) + b_current))
        m_gradient += -(2/n) * x * (y - ((m_current * x) + b_current))
        
    gradient_size = (b_gradient ** 2 + m_gradient ** 2) ** 0.5    
    new_b = b_current - (learning_rate * b_gradient)
    new_m = m_current - (learning_rate * m_gradient)
    
    return [new_b, new_m, gradient_size]

def gradient_descent_runner(points, starting_b, starting_m, num_iterations, learning_rate, printar_rss, plot, tamanho_gradiente_parada):
    b = starting_b
    m = starting_m
    
    iteracao_rss_gd = {"iteracao": [], 
                    "rss": [],
                    "tamanho_gradiente": []}
    
    tamanho_gradiente_anterior = sys.maxint
    i = 0
    while True:
        points_array = array(points)
        b, m, gradient_size = step_gradient(b, m, points_array, learning_rate)
        
        if printar_rss:
            print "iteracao: " + str(i + 1) + ", b:  " + str(b) + ", m: " + str(m) + ", erro: " + \
            str(compute_error_for_given_points(b, m, points_array)) + ", diferença gd: " + \
            str(tamanho_gradiente_anterior - gradient_size)
        
        iteracao_rss_gd["iteracao"].append(i + 1)
        iteracao_rss_gd["rss"].append(compute_error_for_given_points(b, m, points_array))
        iteracao_rss_gd["tamanho_gradiente"].append(gradient_size)
        
        if tamanho_gradiente_anterior - gradient_size < tamanho_gradiente_parada:
            break
        tamanho_gradiente_anterior = gradient_size
        i += 1
    
    if plot:
        return [b, m, iteracao_rss_gd]
    else:
        return [b, m]
    
def run(file, num_iterations = 1000, learning_rate = 0.0001, tamanho_gradiente_parada = 0.0001, printar_rss = False, plot = False):
    points = genfromtxt(file, delimiter=",")
    
    # y = mx + b (fórmula de inclinação da reta)
    initial_b = 0
    initial_m = 0
    
    return gradient_descent_runner(points, initial_b, initial_m, num_iterations, learning_rate, printar_rss, plot, tamanho_gradiente_parada)
    
[b, m, iteracao_rss] = run("data/income.csv", plot = True)

data_iteracao_rss = pd.DataFrame(iteracao_rss, columns = ["iteracao", "rss", "tamanho_gradiente"])

alt.Chart(data_iteracao_rss).mark_point().encode(
    x='iteracao',
    y='tamanho_gradiente'
)
```

O valor encontrado foi com a diferença de tamanho de gradientes de 0,000012, com o learning rate de 0,003.

## Verificando a performance com equações normais

```python
def equacoes_normais(file):
    points = genfromtxt(file, delimiter=",")
    points_array = array(points)
    
    media_x = reduce(lambda a, b: a + b, map(lambda element: element[0], points_array)) / float(len(points_array))
    media_y = reduce(lambda a, b: a + b, map(lambda element: element[1], points_array)) / float(len(points_array))
    
    w1 = reduce(lambda a, b: a + b, map(lambda (x, y): (x - media_x) * (y - media_y), points_array)) / \
        reduce(lambda a, b: a + b, map(lambda (x, y): (x - media_x) ** 2, points_array))
    w0 = media_y - w1 * media_x
    
    return [w0, w1]
```

```python
start_equacoes_normais = datetime.now()
[w0, w1] = equacoes_normais("data/income.csv")
end_equacoes_normais = datetime.now()

start_gradiente_descendente = datetime.now()
[b, m] = run("data/income.csv", learning_rate = 0.003, tamanho_gradiente_parada = 0.000012)
end_gradiente_descendente = datetime.now()

print "tempo de execução equações normais: " + \
    str((end_equacoes_normais - start_equacoes_normais).total_seconds()) + \
    " segundos, modelo encontrado: %.3f*x + %.3f" % (m, b)

print "tempo de execução gradiente descendente: " + \
    str((end_gradiente_descendente - start_gradiente_descendente).total_seconds()) + \
    " segundos, modelo encontrado: %.3f*x + %.3f" % (w1, w0)
```

Como visto acima, o tempo de execução do algoritmo de regressão linear por gradiente descendente demorou cerca de 724 vezes mais lento que o que utiliza as equações normais.
